```python
import numpy as np
import pandas as pd
```

```python
songs = pd.read_csv('500ssongs.csv',encoding='latin1')
```

```python
#title
#description
#appears on
#artist
#released
#position

songs = songs[['title','description','appears on','artist','released']]
```

```python
songs['appears on'] = songs['appears on'].apply(lambda x:x.split())
songs['description'] = songs['description'].apply(lambda x:x.split())
songs['artist'] = songs['artist'].apply(lambda x:x.split())
songs['released'] = songs['released'].apply(lambda x:x.split())
```

```python
songs['appears on'] = songs['appears on'].apply(lambda x:[i.replace(" ","")for i in x])
songs['description'] = songs['description'].apply(lambda x:[i.replace(" ","")for i in x])
songs['artist'] = songs['artist'].apply(lambda x:[i.replace(" ","")for i in x])
songs['released'] = songs['released'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
songs['tags'] = songs['description'] + songs['appears on'] + songs['artist'] + songs['released']
```

```python
new_df = songs[['title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similar[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def rec(songs):
    song_index = new_df[new_df['title'] == songs].index[0]
    distances = similar[song_index]
    songs_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in songs_list:
        print(new_df.iloc[i[0]].title)
```

```python
rec('Imagine')
```
